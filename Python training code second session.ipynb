String Lists Solutions,Ask the user for a string and print out whether this string is a palindrome or not. A palindrome is a string that reads the same forwards and backwards.Exercise 6




```python
wrd = input("Enter a word : ")
if wrd == wrd[::-1]:
  print(wrd + " is a palindrome")
else:
  print(wrd + " is not a palindrome")
```

A sample solution using for loops


```python
def reverse(word):
  x = ""
  for i in range(len(word)):
    x += word[len(word)-1-i]
  return x

word = input("Enter a word : ")
if word == reverse(word):
  print(word + " is a palindrome")
else:
  print(word + " is not a palindrome")
```

Rock Paper Scissors Solutions,Make a two-player Rock-Paper-Scissors game. Hint: Ask for player plays (using input), compare them, print out a message of congratulations to the winner, and ask if the players want to start a new game.Exercise 4




```python
print('''Please pick one:
            rock
            scissors
            paper''')

while True:
    game_dict = {'rock': 1, 'scissors': 2, 'paper': 3}
    player_a = str(input("Player a: "))
    player_b = str(input("Player b: "))
    a = game_dict.get(player_a)
    b = game_dict.get(player_b)
    dif = a - b

    if dif in [-1, 2]:
        print('player a wins.')
        if str(input('Do you want to play another game, yes or no?\n')) == 'yes':
            continue
        else:
            print('game over.')
            break
    elif dif in [-2, 1]:
        print('player b wins.')
        if str(input('Do you want to play another game, yes or no?\n')) == 'yes':
            continue
        else:
            print('game over.')
            break
    else:
        print('Draw.Please continue.')
        print('')
```

Guessing Game One,Ask the user to guess the number, then tell them whether they guessed too low, too high, or exactly right.Exercise 5

```python
import random

status = True
while status:
  i = random.randint(1,9) # Generate a new random number for each game
  print("I am thinking of a number between 1 and 9.")
  while True:
    try:
      guess = int(input("Guess the number: "))
      break # Exit the inner loop if input is valid
    except ValueError:
      print("Invalid input. Please enter an integer.")

  if i == guess:
    print("You are Correct! Right answer!")
  elif guess >= i + 3:
    print("Guess too high")
  elif guess <= i - 3:
    print("Guess too Low")
  else:
    print("Almost Right!")
    print(f"The correct number was: {i}")

  check = input("Want to exit? Type 'exit': ").lower()
  if check == "exit":
    status = False
  print("\n") # Add a newline for better readability between games
```

List Overlap Comprehensions,write a program that returns a list that contains only the elements that are common between the lists (without duplicates). Make sure your program works on two lists of different sizes.Exercise 6

```python
import random

a = random.sample(range(100),20)
print(a)
b = random.sample(range(100),10)
print(b)
nlist = []
nlist.append(print([aa for aa in a if aa in b]))
```
