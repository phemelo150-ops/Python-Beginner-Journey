List Ends,Write a program that takes a list of numbers (for example, a = [5, 10, 15, 20, 25]) and makes a new list of only the first and last elements of the given list.Exersice 2

```python
import random

def get_random_list():
  list_numbers = list(random.sample(range(30),random.randint(2,10)))
  return list_numbers

def take(numbers):
  # This function should return a new list with only the first and last elements
  return [numbers[0], numbers[-1]]

# Generate a random list
my_list = get_random_list()
print("Original list:", my_list)

# Get the first and last elements using the take function
first_and_last = take(my_list)
print("First and last elements:", first_and_last)
```

Reverse Word Order,Write a program (using functions!) that asks the user for a long string containing multiple words. Print back to the user the same string, except with the words in backwards order.Exercise 5

```python
def reverse(lister):
  spliter = lister.split()
  re = list(reversed(spliter))
  joiner = " ".join(re) # Added a space for better readability when joining
  return joiner

string = "My name is Phemelo"
print(reverse(string))
```

```python
test = "My name is Phemelo"
def backwards_string(input_string):
  split = input_string.split()
  x = []
  for i in range(len(split)-1, -1, -1):
    x.append(split[i])
  return " ".join(x)

print(backwards_string(test))
```
